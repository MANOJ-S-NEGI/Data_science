#### AND|OR

##### AND combines two or more conditions for a single query. All of the conditions used with this operator must be met in order to show the results.

##### OR is used similarly, but it will output result with rows that comply with either of the conditions.

```python
mydb=connection.connect(host='localhost',user='root',passwd='root',use_pure=True)
cursor=mydb.cursor()
cursor.execute("select * from shop.item_list where product_id=9")
fetch=cursor.fetchall()
print(fetch,"ok")
    
```

```python
## AND 
mydb=connection.connect(host='localhost',user='root',passwd='root',use_pure=True)
cursor=mydb.cursor()
cursor.execute("select * from shop.item_list where price=25000 and quantity=1")
fetch=cursor.fetchall()
print(fetch,"ok")
```

#### CREATE VIEW
##### CREATE VIEW creates a narrower version of an existing table by getting a set of results based on a certain query. A view is not much different from a real table: it contains columns and rows with data, but it doesn't contain the fields of the real table that are irrelevant for your particular purpose.


CREATE VIEW  any name for new list  AS
SELECT table_col1, table_col_2
FROM table
WHERE data = condition;

```python
mydb=connection.connect(host='localhost',database="shop",user='root',passwd='root',use_pure=True)
cursor=mydb.cursor()
cursor.execute("CREATE VIEW new_item_list1 AS select product_name,price from item_list where price between 3000 and 25000")
fetch=cursor.fetchall()
print(fetch,"\n query executed:ok")
```

#### GROUP BY
##### Combine GROUP BY with SELECT statement in order to arrange identical data (rows with the same value) into groups (summarizing rows).

```python
mydb=connection.connect(host='localhost',database="shop",user='root',passwd='root',use_pure=True)
cursor=mydb.cursor()
cursor.execute(" SELECT product_name,price FROM item_list GROUP BY product_name;")
group=cursor.fetchall()
print(group,"\n query executed:ok")
```

#### HAVING
##### HAVING specifies that you need to filter the results to only the rows that fulfill the described condition.

###### It performs the same action as the WHERE clause. The difference is that HAVING is used only for aggregate functions as WHERE doesn't work with them.

```python
mydb=connection.connect(host='localhost',database="shop",user='root',passwd='root',use_pure=True)
cursor=mydb.cursor()
cursor.execute("SELECT COUNT(product_id), product_name FROM item_list GROUP BY price HAVING COUNT(product_id) > 2;")
having=cursor.fetchall()
print("\n query executed:ok")
```

```python
for h in having:
    print(h)
```

#### Order by:

#### Inner_join 

###### INNER JOIN combines rows from different tables.

```python
mydb=connection.connect(host='localhost',database="shop",user='root',passwd='root',use_pure=True)
cursor=mydb.cursor()
cursor.execute('SELECT shop.item_list.product_id,shop.item_list.product_name, shop.item_list_loc.Location FROM shop.item_list_loc INNER JOIN shop.item_list ON shop.item_list.product_id = shop.item_list_loc.product_id')
join=cursor.fetchall()
print("\n query executed:ok")
```

```python
for j in join:
    print(j)
```

#### Left Join:
###### LEFT JOIN retrieves records from the left table that match records in the right table. Some databases have a slightly different statement for this – LEFT OUTER JOIN.

```python
mydb=connection.connect(host='localhost',database="shop",user='root',passwd='root',use_pure=True)
cursor=mydb.cursor()
cursor.execute('SELECT shop.item_list.product_id,shop.item_list.product_name, shop.item_list_loc.Location FROM shop.item_list_loc  left JOIN shop.item_list ON shop.item_list.product_id = shop.item_list_loc.product_id ORDER BY item_list.product_Name')
ljoin=cursor.fetchall()
print("query executed:ok")
```

```python
# arrange sequence by name
for l in ljoin:
    print(l)
```
